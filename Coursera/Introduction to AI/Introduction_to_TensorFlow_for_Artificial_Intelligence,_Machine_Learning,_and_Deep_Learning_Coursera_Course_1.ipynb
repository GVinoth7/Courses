```
import tensorflow as tf

(train_img,train_lbl), (test_img, test_lbl) = tf.keras.datasets.mnist.load_data()
```

```
train_img = train_img.reshape(train_img.shape[0], 28, 28, 1) / 255.0
test_img = test_img.reshape(test_img.shape[0], 28, 28, 1)/ 255.0

model = tf.keras.Sequential()
model.add(tf.keras.layers.Conv2D(32, (5,5), input_shape = (28, 28, 1), activation = 'relu'))
model.add(tf.keras.layers.MaxPooling2D(2, 2))
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(256, activation ='relu'))
model.add(tf.keras.layers.Dense(128, activation = 'relu'))
model.add(tf.keras.layers.Dense(64, activation = 'relu'))
model.add(tf.keras.layers.Dense(10, activation = 'softmax'))
```

```
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```
class Mycallbacks(tf.keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs = {}):
        if logs.get('acc') > 0.998:
            self.model.stop_training = True
            print('\n Reached 99.8% accuracy so cancelling training!')
callback = Mycallbacks()

history = model.fit(train_img, train_lbl, validation_data = (test_img, test_lbl), epochs = 20, callbacks = [callback])
            
```
